```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
tips = pd.read_csv(url)
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python
sns.set_style('whitegrid')
sns.axes_style('whitegrid')
sns.violinplot(x = 'total_bill', data = tips, color = 'r', saturation = 0.3)
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tips, order = ['Thur','Fri','Sat','Sun'])
plt.figure(figsize = (10,6))
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tips, order = ['Thur','Fri','Sat','Sun'], split = True, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tips, order = ['Thur','Fri','Sat','Sun'], split = True, hue = 'smoker', palette = 'Set1')
```
