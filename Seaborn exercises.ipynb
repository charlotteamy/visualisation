```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

df1 = pd.read_csv('Pokemon.csv',index_col = 0, encoding = 'unicode_escape')
```

```python
df1.head()
```

```python
df1.shape
```

```python
sns.lmplot(x = 'Attack', y = 'Defense', data = df1, hue = 'Stage', fit_reg = False)
```

```python
df2 = df1.drop(['Total','Stage','Legendary'], axis = 1)
```

```python
sns.boxplot(data = df2)
sns.set_style('whitegrid')
```

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data = df2)
```

```python
plt.figure(figsize = (10,6))

sns.violinplot(x = 'Type 1', y = 'Attack', data = df2)
sns.swarmplot(x = 'Type 1', y = 'Attack', data = df2, color = 'k', alpha = 0.7)
plt.title('Attack by Type')
```
